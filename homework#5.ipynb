First start by throwing a signal-free background. For Problem 1 choose a Normal distribution with some modest $\sigma = 2$. Create a million background events.

```python

%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
import statistics
import math
from scipy import stats
import h5py
import pandas as pd
```

```python
plt.rcParams["figure.figsize"] = (7,5)
```

```python
size = 1000000
bkgd = np.random.normal(0, 2, size)
signal = np.random.uniform(0, 20, size)
data = bkgd + signal

plt.hist(bkgd, bins=100, density=True)
plt.title('Background')
plt.ylabel('Probability Density')
plt.xlabel('Measurement')
plt.show()

plt.hist(signal, bins=100, density=True)
plt.title('Signal')
plt.ylabel('Probability Density')
plt.xlabel('Measurement')
plt.show()

plt.hist(data, bins=100, density=True)
plt.title('Data')
plt.ylabel('Probability Density')
plt.xlabel('Measurement')
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (13,10)
```

```python
signaledges = np.linspace(0,20,40)
dataedges = np.linspace(-7,27,68)

Psd, temp, temp2= np.histogram2d(data,signal, bins=[dataedges,signaledges], density=True)

datacenters = (dataedges[:-1] + dataedges[1:]) / 2
signalcenters = (signaledges[:-1] + signaledges[1:]) / 2
h = plt.pcolormesh(datacenters,signalcenters,Psd.T)
plt.colorbar(h)
plt.title('Observed Data vs True Signal', fontsize=24)
plt.ylabel('True signal, $P(s|d)$', fontsize = 24)
plt.xlabel('Observed data, $P(d|s)$', fontsize = 24)
plt.show()
```
