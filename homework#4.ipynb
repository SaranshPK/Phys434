## Problem 0: some fake data to mess with

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
import pandas as pd
from scipy import stats
```

```python
t = np.linspace(0,np.pi*100,np.int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=np.int(1e5))*np.sqrt(N)+N
```

```python
plt.hist(bkgd, 300, density=True)
plt.title('Background Distribution')
plt.xlabel('Background')
plt.ylabel('Density')
plt.show()
```

## Problem 1: subsets of data

__1.a.__

```python
plt.scatter(t[:1000], bkgd[:1000])
plt.title('Background from 0 to 1000 data points')
plt.xlabel('Time')
plt.ylabel('Background')
plt.show()
```

__1.b.__

```python
def plotPoints(start, count, time, background):
    plt.scatter(time[start:start+count], background[start:start+count])
    plt.title(f'Background from {start} to {start+count} data points')
    plt.xlabel('Time')
    plt.ylabel('Background')
    plt.show()
```

```python
plotPoints(2000, 2000, t, bkgd)
```

__1.b.__

```python
plt.scatter(t[::1000], bkgd[::1000])
plt.title('Background every 1000 data points')
plt.xlabel('Time')
plt.ylabel('Background')
plt.show()
```

```python
plt.hist2d(t, bkgd, bins = 40)
plt.title('Density of Background Values')
plt.xlabel('Time')
plt.ylabel('Background')
cbar = plt.colorbar()
cbar.set_label('Number of Data Points')
plt.show()
```

__2.b.__ This plot shows the distribution of the background data across time. As we can see, at any given time, the background is most dense around 10 and least dense far away from 10. The color represents this density, with yellow being the most dense and violet being the least dense.
